```python
#calcuration libraries
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

滑らかな管の管摩擦係数p161

```python
L = 50
d = 0.2
A = np.pi * d**2 / 4
rho = 1000
mu = 0.001
nu = 1.0 * 10**(-6)
Q = np.linspace(0.05,10,)
U = Q * 10**(-3) / A
Re = U * d / nu
#print(Re)
```

Laminar

```python
#Darcy–Weisbach Equation Re < 2300
lamda_lam = 64 / Re
dp_lam = lamda_lam * (L/d) * (rho * U**2 / 2)
```

Turbulent

```python
#Blasius Equation Re < 10**5
lamda_Blasius = 0.3164 * Re**(-1/4)
dp_Blasius = lamda_Blasius * (L/d) * (rho * U**2 / 2)
```

```python
#Nikuradse Equation　Re > 10**5
lamda_Nikuradse = 0.032 + 0.211 * Re**(-0.23)
dp_Nikuradse = lamda_Nikuradse * (L/d) * (rho * U**2 / 2)
```

```python
#Prandtl-Karman Equation
```

```python
plt.plot(Q,dp_lam)
plt.plot(Q,dp_Blasius)
plt.plot(Q,dp_Nikuradse)
plt.xlabel(r'Flow rate Q$\,\mathrm{[l/s]}$')
plt.ylabel(r'Pressure loss $\Delta$P$\,\mathrm{[Pa]}$')
```

粗い管の管摩擦係数p163

```python
epsilon = 0.01 #roughness coefficient
y = np.linspace(0.0001,0.005,)
U_ast = 5.75 * np.log(y/epsilon) + 8.5
plt.xlim([0,5])
plt.plot(U_ast, y)
plt.xlabel(r'valocity distribution$\,\mathrm{[-]}$')
plt.ylabel(r'y$\,\mathrm{[m]}$')
```
